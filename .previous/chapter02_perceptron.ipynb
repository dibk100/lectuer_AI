#### 1.What is **Perceptron**
퍼셉트론은 다수의 신호를 입력 받아 하나의 신호를 출력한다.   
퍼셉트론 신호는 흐름을 만들고 정보를 앞으로 전달한다.   

x1,x2는 입력 신호, y는 출력 신호, w1,w2는 가중치.   
*x1,x2,y(원형)은 뉴런or노드 라고 한다.   
   
입력 신호가 뉴련(y)에 보내질 때는 고유한 가중치(w)가 곱해서 전달되는데(x1w1, x2w2)   
신호의 총 합이 정해진 한계(임계값)를 넘어설 때만 1을 출력한다.(= 뉴련이 활성화한다)   

> 퍼셉트론 동작 원리는 위에 내용이 전부이다.

#### 2.단순한 논리 회로   
논리연산자와 퍼셉트론 : AND, OR,NAND

#### 3.퍼셉트론 구현하기
AND게이트

```python
def function_and(x1,x2):
    w1,w2,theta = 0.5,0.5,0.7
    formula = x1*w1+x2*w2
    
    if formula <= theta :
        return 0
    else :
        return 1

function_and(1,1)
# output : 1
```

> Note :   
- 가중치 w1,w2는 입력 신호의 중요도를 나타내고,
- 임계값(theta : θ : 편향)는 뉴련이 활성화되는 기준을 나타낸다.
    - 임계값이 높으면 분류기준이 높아지고(뉴련활성화 기준이 높아짐)
    - 높을수록 엄격히 분류되기 때문에 모델이 간단해지는 경향이 있다.(underfitting)

> 퍼셉트론 학습 방법 :   

처음에는 임의의 weight(가중치)로 시작한다.  
학습 데이터를 퍼셉트론 모형에 입력해서 분류가 잘못됐을 때 weight를 개선해나간다.  
학습하면서 weight를 계속 조정(adjust)되는 것이다.


#### 4.퍼셉트론의 한계점   
AND,OR,NAND Gate만 가능하고(선형분류), XOR 같은 데이터 형태는 분류가 불가능

#### 5.다층 퍼셉트론   
퍼셉트론의 층을 쌓아 다층 퍼셉트론(multi-layer perceptron)을 만들어 비선형 영역을 분류할 수 있다.   
즉, AND,OR,NAND를 여러 층 쌓아서 XOR를 구현할 수 있다.

S1,S2는 은닉층    
x1,x2는 입력값

> XOR 구현하기

```python
class Perceptron:
    w1,w2,theta = 0.5,0.5,0.7
    or_theta = 0.2
    
    def and_Gate(self,x1,x2) -> int :
        formula = x1*self.w1+x2*self.w2
        
        if formula <= self.theta :
            return 0
        else :
            return 1
    
    # and gate에서 theta와 가중치의 부호만 다름
    def nand_Gate(self,x1,x2) -> int :
        formula = x1*(self.w1*-1)+x2*(self.w2*-1)
    
        if formula <= (self.theta*-1) :
            return 0
        else :
            return 1
    
    # and gate에서 theta만 다름
    def or_Gate(self,x1,x2) -> int :
        formula = x1*self.w1+x2*self.w2
    
        if formula <= self.or_theta :
            return 0
        else :
            return 1
        
    def xor_Gate(self,x1,x2) -> int :
        s1 = self.nand_Gate(x1,x2)
        s2 = self.or_Gate(x1,x2)
        
        return self.and_Gate(s1,s2)
```
